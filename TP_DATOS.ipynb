```python
import pandas as pd

# Le pedimos a la computadora que lea tu archivo de Excel.
# ATENCIÓN: Cambiá "EL_NOMBRE_DE_TU_ARCHIVO.xlsx" por el nombre exacto de tu archivo subido.
df = pd.read_excel("/content/Censo 2022 - Vicente López-modif.xlsx")

# Le pedimos la radiografía del archivo (el diagnóstico)
df.info()
```

```python
import pandas as pd
import numpy as np

# =====================================================================
# TAREA 1 Y 2: CORREGIR TIPOS DE DATOS (Las columnas tramposas)
# =====================================================================
df['Perforación con bomba manual'] = pd.to_numeric(df['Perforación con bomba manual'], errors='coerce').fillna(0).astype(int)
df['Pozo sin bomba'] = pd.to_numeric(df['Pozo sin bomba'], errors='coerce').fillna(0).astype(int)


# =====================================================================
# TAREA 3: SEPARAR LA GEOLOCALIZACIÓN (Latitud y Longitud)
# =====================================================================
if 'Geolocalización' in df.columns:
    coordenadas = df['Geolocalización'].str.split(',', expand=True)
    df['Latitud'] = pd.to_numeric(coordenadas[0], errors='coerce')
    df['Longitud'] = pd.to_numeric(coordenadas[1], errors='coerce')
    df.drop(columns=['Geolocalización'], inplace=True)


# =====================================================================
# TAREA 4: AGRUPAR LAS EDADES (Con los nuevos nombres de rangos)
# =====================================================================
# 1. Grupo de 0 a 15 años (reales: 0 a 14)
df['Poblacion_0_a_15'] = df['0-4'] + df['5-9'] + df['10-14']

# 2. Grupo de 16 a 65 años (reales: 15 a 64)
df['Poblacion_16_a_65'] = (
    df['15-19'] + df['20-24'] + df['25-29'] + df['30-34'] + df['35-39'] +
    df['40-44'] + df['45-49'] + df['50-54'] + df['55-59'] + df['60-64']
)

# 3. Grupo de 66 años o más (reales: 65 o más)
df['Poblacion_66_o_mas'] = (
    df['65-69'] + df['70-74'] + df['75-79'] + df['80-84'] +
    df['85-89'] + df['90-94'] + df['95-99'] + df['100 y más']
)

# Borramos las 21 columnas viejas para limpiar la tabla
columnas_edades_viejas = [
    '0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39',
    '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74',
    '75-79', '80-84', '85-89', '90-94', '95-99', '100 y más'
]
df.drop(columns=columnas_edades_viejas, inplace=True)


# =====================================================================
# CONTROL FINAL: Ver cómo quedó nuestra nueva tabla limpia
# =====================================================================
df.info()
```
